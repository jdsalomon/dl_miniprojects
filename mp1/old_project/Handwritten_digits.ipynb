```python
# Helper functions

# Plot Accuracy and Loss of Model
def plot_accuracy_loss(train_accuracy, train_loss, test_accuracy, test_loss):
    f = plt.figure(figsize=(15, 5))
    ax1 = f.add_subplot(121)
    ax2 = f.add_subplot(122)
    ax1.title.set_text("Accuracy over epochs")
    ax2.title.set_text("Loss over epochs")
    ax1.plot(train_accuracy, c='b', label='Train')
    ax1.plot(test_accuracy, c='r', label='Test')
    ax2.plot(train_loss, c='b', label='Train')
    ax2.plot(test_loss, c='r', label='Test')
    plt.legend()
```

### Final output recognition

```python
class Sub_LeNet5(nn.Module):
    def __init__(self, lr=LEARNING_RATE):
        super(Sub_LeNet5, self).__init__()        
        
        #First image layers
        self.conv1 = nn.Conv2d(1, 16, kernel_size=5, stride=1)
        self.average = nn.AvgPool2d(2, stride=2)
        self.conv2 = nn.Conv2d(16, 120, kernel_size=5, stride=1)
                
        self.fc1 = nn.Linear(120, 82)
        self.fc2 = nn.Linear(82, 10)
                
        #Second image layer
        self.conv1_2 = nn.Conv2d(1, 16, kernel_size=5, stride=1)
        self.average_2 = nn.AvgPool2d(2, stride=2)
        self.conv2_2 = nn.Conv2d(16, 120, kernel_size=5, stride=1)
        self.fc1_2 = nn.Linear(120, 82)
        self.fc2_2 = nn.Linear(82, 10)
        
        #The functions are the same, but they are different functions. This avoids weight sharing.
        
        #Digit Comparison layer
        self.fc3 = nn.Linear(20,100)
        self.fc4 = nn.Linear(100,100)
        self.fc5 = nn.Linear(100,1)
        
        
        self.prev_loss = -1
        self.lr_min = lr / 100

        self.best_test_acc = -1
        self.best_epoch = -1
        
    def forward(self, im1, im2):
        im1 = im1.reshape(-1, 1, 14, 14)
        out1 = torch.tanh(self.conv1(im1))
        out1 = self.average(out1)
        out1 = torch.tanh(self.conv2(out1))
        out1 = out1.view(-1, out1.shape[1])
        out1 = torch.relu(self.fc1(out1))
        out1 = torch.log_softmax(self.fc2(out1), dim=1)
        
        im2 = im2.reshape(-1, 1, 14, 14)
        out2 = torch.tanh(self.conv1_2(im2))
        out2 = self.average_2(out2)
        out2 = torch.tanh(self.conv2_2(out2))
        out2 = out2.view(-1, out2.shape[1])
        out2 = torch.relu(self.fc1_2(out2))
        out2 = torch.log_softmax(self.fc2_2 (out2), dim=1)
     
        pair = torch.cat((out1,out2),1)
        pair = torch.relu(self.fc3(pair))
        pair = torch.relu(self.fc4(pair))
        comparison = torch.sigmoid(self.fc5(pair))
        return comparison
    
    def update_lr(self, new_loss, old_loss, optimizer, threshold=0.03, verbose=False):
        for param_group in optimizer.param_groups:
            diff = old_loss - new_loss
            if diff/old_loss <= threshold and param_group['lr'] > self.lr_min:
                param_group['lr'] = param_group['lr'] / 2

                if verbose:
                    print("Lower the rate. New rate :", param_group['lr'])
                
    def fit(self, train, train_target, test, test_target, loss_function, optimizer, batch_size=BATCH_SIZE, adaptive_lr=True, suffix="", verbose=False):
        train_target = train_target.float()
        train_acc, train_losses, test_acc, test_losses = [], [], [], []
        
        for epoch in range(N_EPOCHS):
            sum_loss, correct, total = 0, 0, 0
            permutation = torch.randperm(train.size(0))
            
            for batch_start in range(0, train.size(0), batch_size):
                # Get the batch
                indices = permutation[batch_start:batch_start+batch_size]
                batch_train, batch_target = train[indices], train_target[indices]
                batch_train1, batch_train2 = batch_train[:, 0], batch_train[:, 1]

                # Forward pass                
                output = self(batch_train1, batch_train2).view(-1)
                # Metrics
                loss = loss_function(output, batch_target)
                sum_loss += loss.item()
                
                pred = output.round()
                total += batch_target.size(0)
                correct += (batch_target == pred).sum().item()
                
                # Backward pass
                optimizer.zero_grad()
                loss.backward()
                optimizer.step()
            
            if adaptive_lr:
                self.update_lr(sum_loss, self.prev_loss, optimizer)
                self.prev_loss = sum_loss
                
            # Test the model
            #test_loss = copy.deepcopy(loss_function)
            test_pred, acc_test, test_loss = self.predict(test, test_target, loss_function)
            if acc_test > self.best_test_acc:
                self.best_test_acc = acc_test
                self.best_epoch = epoch
                #torch.save(best_model.state_dict(), f'data/Sub_LeNet5/best_sub_LN5{suffix}.pt')
                
            # Save the accuracy and loss
            train_acc.append(correct/total)
            train_losses.append(sum_loss)
            test_acc.append(acc_test)
            test_losses.append(test_loss)
            
            if verbose:
                for param_group in optimizer.param_groups:
                    print(f'Epoch: {epoch}', 
                          'Train aux loss: {:.5f}'.format(train_losses[-1]), 
                          f'Train Accuracy: {train_acc[-1]}', 
                          f'Test Accuracy: {test_acc[-1]}',
                          f'Learning Rate: {param_group["lr"]}',
                          sep='\t')
        if verbose:
            print(f'Best Model was at epoch {self.best_epoch} and has test accuracy of {self.best_test_acc}')
        
        #torch.save(best_model.state_dict(), f'Project 1/data/Compare_With_LeNet5/best_CWLN5{suffix}.pt')
    
        return train_acc, train_losses, test_acc, test_losses
                        
    def predict(self, test, target, loss_function):
        target = target.float()
        test1, test2 = test[:, 0], test[:, 1]
        with torch.no_grad():
            output = self(test1, test2).view(-1)
            loss = loss_function(output, target)
            pred = output.round()
            correct = (target == pred).sum().item()
            acc = correct / test.size(0)
            return pred, acc, loss
```

### Final Output recognition using weight sharing

```python
class Sub_LeNet5_WS(nn.Module):
    def __init__(self, lr=LEARNING_RATE):
        super(Sub_LeNet5_WS, self).__init__()
        
        #Image recognition functions - Similar to previous model, but the same functions are used for the images to apply weight sharing on those functions.
        self.conv1 = nn.Conv2d(1, 16, kernel_size=5, stride=1)
        self.average = nn.AvgPool2d(2, stride=2)
        self.conv2 = nn.Conv2d(16, 120, kernel_size=5, stride=1)
                
        self.fc1 = nn.Linear(120, 82)
        self.fc2 = nn.Linear(82, 10)
        
        #Digit Comparison layer
        self.fc3 = nn.Linear(20,100)
        self.fc4 = nn.Linear(100,100)
        self.fc5 = nn.Linear(100,1)
        
        #For adaptive lr
        self.prev_loss = -1
        self.lr_min = lr / 100
        
        #To find the best model (to avoid overfitting)
        self.best_test_acc = -1
        self.best_epoch = -1
        
    def forward(self, im1, im2):   
        im1 = im1.reshape(-1, 1, 14, 14)
        out1 = torch.tanh(self.conv1(im1))
        out1 = self.average(out1)
        out1 = torch.tanh(self.conv2(out1))
        out1 = out1.view(-1, out1.shape[1])
        out1 = torch.relu(self.fc1(out1))
        out1 = torch.log_softmax(self.fc2(out1), dim=1)
        
        im2 = im2.reshape(-1, 1, 14, 14)
        out2 = torch.tanh(self.conv1(im2))
        out2 = self.average(out2)
        out2 = torch.tanh(self.conv2(out2))
        out2 = out2.view(-1, out2.shape[1])
        out2 = torch.relu(self.fc1(out2))
        out2 = torch.log_softmax(self.fc2(out2), dim=1)
     
        pair = torch.cat((out1,out2),1)
        pair = torch.relu(self.fc3(pair))
        pair = torch.relu(self.fc4(pair))
        comparison = torch.sigmoid(self.fc5(pair))
        return comparison
    
    def update_lr(self, new_loss, old_loss, optimizer, threshold=0.03):
        for param_group in optimizer.param_groups:
            diff = old_loss - new_loss
            if diff/old_loss <= threshold and param_group['lr'] > self.lr_min:
                param_group['lr'] = param_group['lr'] / 2

                if verbose:
                    print("Lower the rate. New rate :", param_group['lr'])
        
    def fit(self, train, train_target, test, test_target, loss_function, optimizer, batch_size=BATCH_SIZE, adaptive_lr=True, suffix="", verbose=False):
        train_target = train_target.float()
        train_acc, train_losses, test_acc, test_losses = [], [], [], []
        
        for epoch in range(N_EPOCHS):
            sum_loss, correct, total = 0, 0, 0
            permutation = torch.randperm(train.size(0))
            
            for batch_start in range(0, train.size(0), batch_size):
                # Get the batch
                indices = permutation[batch_start:batch_start+batch_size]
                batch_train, batch_target = train[indices], train_target[indices]
                batch_train1, batch_train2 = batch_train[:, 0], batch_train[:, 1]

                # Forward pass                
                output = self(batch_train1, batch_train2).view(-1)
                
                # Metrics
                loss = loss_function(output, batch_target)
                sum_loss += loss.item()
                
                pred = output.round()
                total += batch_target.size(0)
                correct += (batch_target == pred).sum().item()
                
                # Backward pass
                optimizer.zero_grad()
                loss.backward()
                optimizer.step()
            
            if adaptive_lr:
                self.update_lr(sum_loss, self.prev_loss, optimizer)
                self.prev_loss = sum_loss
                
            # Test the model
            test_pred, acc_test, test_loss = self.predict(test, test_target, loss_function)
            if acc_test > self.best_test_acc:
                self.best_test_acc = acc_test
                self.best_epoch = epoch
                #torch.save(best_model.state_dict(), f'Project 1/data/Sub_LeNet5_WS/best_sub_LN5_ws{suffix}.pt')
                
            # Save the accuracy and loss
            train_acc.append(correct/total)
            train_losses.append(sum_loss)
            test_acc.append(acc_test)
            test_losses.append(test_loss)
            
            if verbose:
                for param_group in optimizer.param_groups:
                    print(f'Epoch: {epoch}', 
                          'Train aux loss: {:.5f}'.format(train_losses[-1]), 
                          f'Train Accuracy: {train_acc[-1]}', 
                          f'Test Accuracy: {test_acc[-1]}',
                          f'Learning Rate: {param_group["lr"]}',
                          sep='\t')
        if verbose:
            print(f'Best Model was at epoch {self.best_epoch} and has test accuracy of {self.best_test_acc}')
        
    
        return train_acc, train_losses, test_acc, test_losses
                        
    def predict(self, test, target, loss_function):
        target = target.float()
        with torch.no_grad():
            test1, test2 = test[:, 0], test[:, 1]
            
            output = self(test1, test2).view(-1)
            loss = loss_function(output, target)
            pred = output.round()
            correct = (target == pred).sum().item()
            acc = correct / test.size(0)
            return pred, acc, loss
```

### Digit recognition using auxiliary loss and weight sharing

```python
class Sub_LeNet5_WS_auxLoss(nn.Module):
    def __init__(self, lr=LEARNING_RATE):
        super(Sub_LeNet5_WS_auxLoss, self).__init__()        
        self.conv1 = nn.Conv2d(1, 16, kernel_size=5, stride=1)
        self.average = nn.AvgPool2d(2, stride=2)
        self.conv2 = nn.Conv2d(16, 120, kernel_size=5, stride=1)
                
        self.fc1 = nn.Linear(120, 82)
        self.fc2 = nn.Linear(82, 10)
                
        self.prev_loss = -1
        self.lr_min = lr / 100
        
        #To find the best model (to avoid overfitting)
        self.best_test_acc = -1
        self.best_epoch = -1

    def forward(self, im1, im2):
        im1 = im1.reshape(-1, 1, 14, 14)
        out1 = torch.tanh(self.conv1(im1))
        out1 = self.average(out1)
        out1 = torch.tanh(self.conv2(out1))
        out1 = out1.view(-1, out1.shape[1])
        out1 = torch.relu(self.fc1(out1))
        out1 = torch.log_softmax(self.fc2(out1), dim=1)
        
        im2 = im2.reshape(-1, 1, 14, 14)
        out2 = torch.tanh(self.conv1(im2))
        out2 = self.average(out2)
        out2 = torch.tanh(self.conv2(out2))
        out2 = out2.view(-1, out2.shape[1])
        out2 = torch.relu(self.fc1(out2))
        out2 = torch.log_softmax(self.fc2(out2), dim=1)
     
        return out1, out2
    
    def update_lr(self, new_loss, old_loss, threshold=0.03):
        for param_group in optimizer.param_groups:
            diff = old_loss - new_loss
            if diff/old_loss <= threshold and param_group['lr'] > self.lr_min:
                param_group['lr'] = param_group['lr'] / 2

                if verbose:
                    print("Lower the rate. New rate :", param_group['lr'])
                
    def fit(self, train, train_target, train_classes, test, test_target, test_classes, aux_loss_function, optimizer, batch_size=BATCH_SIZE, adaptive_lr=True, verbose=False):
        train_target = train_target.float()
        train_acc, train_losses, test_acc, test_losses = [], [], [], []
    
        for epoch in range(N_EPOCHS):
            sum_loss, correct, total = 0, 0, 0
            permutation = torch.randperm(train.size(0))
            for batch_start in range(0, train.size(0), batch_size):
                # Get the batch
                indices = permutation[batch_start:batch_start+batch_size]
                batch_train, batch_target, batch_classes = train[indices], train_target[indices], train_classes[indices]
                batch_train1, batch_train2 = batch_train[:, 0], batch_train[:, 1]
                batch_class1, batch_class2 = batch_classes[:, 0], batch_classes[:, 1]
                
                # Forward pass                
                output1, output2 = self(batch_train1, batch_train2)
                pred1, pred2 = torch.argmax(output1, 1), torch.argmax(output2, 1)
                
                # Metrics 
                aux_loss_1 = aux_loss_function(output1, batch_class1)
                aux_loss_2 = aux_loss_function(output2, batch_class2)
                aux_loss = aux_loss_1 + aux_loss_2
                
                sum_loss += aux_loss.item()
                pred = (pred1 <= pred2).float()
                total += batch_target.size(0)
                correct += (batch_target == pred).sum().item()
                
                # Backward pass
                optimizer.zero_grad()
                aux_loss.backward()
                optimizer.step()
                
            if adaptive_lr:
                self.update_lr(sum_loss, self.prev_loss)
                self.prev_loss = sum_loss
            # Test the model
            
            test_pred, acc_test, test_loss = self.predict(test, test_target, test_classes, aux_loss_function)
            
            if acc_test > self.best_test_acc:
                self.best_test_acc = acc_test
                self.best_epoch = epoch
                #torch.save(best_model.state_dict(), f'Project 1/data/Sub_LeNet5_WS_auxLoss/best_sub_LN5_ws_aL{suffix}.pt')
                
            # Save the accuracy and loss
            train_acc.append(correct/total)
            train_losses.append(sum_loss)
            test_acc.append(acc_test)
            test_losses.append(test_loss)
            
            if verbose:
                for param_group in optimizer.param_groups:
                    print(f'Epoch: {epoch}', 
                          'Train aux loss: {:.5f}'.format(train_losses[-1]), 
                          f'Train Accuracy: {train_acc[-1]}', 
                          f'Test Accuracy: {test_acc[-1]}',
                          f'Learning Rate: {param_group["lr"]}',
                          sep='\t')
        if verbose:
            print(f'Best Model was at epoch {self.best_epoch} and has test accuracy of {self.best_test_acc}')
                
        return train_acc, train_losses, test_acc, test_losses
                
            
            
    def predict(self, test, target, classes, loss_function):
        target = target.float()
        input1, input2 = test[:, 0], test[:, 1]
        classes1, classes2 = classes[:, 0], classes[:, 1]
        with torch.no_grad():
            output1, output2 = self(input1, input2)
            pred1, pred2 = torch.argmax(output1, 1), torch.argmax(output2, 1)
            loss = loss_function(output1, classes1) + loss_function(output2, classes2)
            pred = (pred1 <= pred2).float()
            correct = (target == pred).sum().item()
            acc = correct / test.size(0)
            return pred, acc, loss
```

### Optimizing Learning rate

#### Find best learning rate by Grid Search

```python
learning_rates = [0.1, 0.05, 0.01, 0.001, 0.0001, 0.00001]
batch_sizes = [50, 100, 250, 500, 1000]

def find_best_lr(results, model_num, verbose=False):
    if verbose:
        print("Accuracy given lr for model {0}".format(model_num))
    best_lr = -1
    best_acc = -1
    for i in range(len(learning_rates)):
        mean = results[:,i,:].mean(1)
        if verbose:
            print("LR = {0} : ".format(learning_rates[i]))
            print("Train acc : ", mean[2 * model_num + 0])
            print("Test acc : ", mean[2 * model_num + 1])
            print()
        if mean[model_num + 1] > best_acc:
            best_acc = mean[2 * model_num + 1]
            best_lr = learning_rates[i]
    return best_lr, best_acc


def find_best_bs(results, model_num, verbose=False):
    if verbose:
        print("Accuracy given batch size for model {0}".format(model_num))
    best_lr = -1
    best_acc = -1
    for i in range(len(batch_sizes)):
        mean = results[:,i,:].mean(1)
        if verbose:
            print("BS = {0} : ".format(batch_sizes[i]))
            print("Train acc : ", mean[2 * model_num + 0])
            print("Test acc : ", mean[2 * model_num + 1])
            print()
        if mean[model_num + 1] > best_acc:
            best_acc = mean[2 * model_num + 1]
            best_lr = batch_sizes[i]
            
    return best_lr, best_acc
```
